DATASET YANG DIGUNAKAN ADALAH TINGKAT PENGANGGURAN DI INDONESIA (DALAM PERSEN) DARI TAHUN 2016-2023 (BULAN AGUSTUS)

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
tingkat_pengangguran = [5.61, 5.5, 5.3, 5.23, 7.07, 6.49, 5.86, 5.32]
tahun = [2016, 2017, 2018, 2019, 2020, 2021, 2022, 2023]

plt.plot(
    tahun,
    tingkat_pengangguran,
    marker = 'x',
)
plt.title('Tingkat Pengangguran Terbuka di Indonesia dari Tahun 2016-2023')
plt.ylabel('Tingkat Pengangguran (%)')
plt.xlabel('Tahun (Periode Agustus)')
plt.show()
```

Berdasarkan grafik diatas, dapat dilihat bahwa tingkat pengangguran di Indonesia sejak tahun 2016 mulai menurun hingga tahun 2019. Namun, pada tahun 2020 tingkat pengangguran di Indonesia meningkat drastis dari sebelumnya, dan merupakan tingkat pengangguran tertinggi selama 4 tahun terakhir. Penyebab utama dari tingginya angka pengganguran pada tahun 2020 adalah karena pandemi covid-19 yang menyebabkan banyak orang kehilangan pekerjaannya dan tidak bisa mencari pekerjaan. Di tahun 2021, tingkat pengangguran sudah mulai menurun kembali, diperkirakan karena pada tahun ini aktivitas sudah mulai berjalan normal. Penurunan tingkat pengangguran ini terus terjadi hingga tahun 2023, kembali menyesuaikan dengan tingkat pengangguran sebelum covid-19 terjadi. Penyebabnya adalah karena pada tahun-tahun tersebut seluruh aktivitas sudah mulai berjalan dengan normal, orang-orang bisa bekerja kembali seperti saat sebelum pandemi covid-19 terjadi, dan mereka yang belum memiliki pekerjaan bisa mendapatkan pekerjaan.
